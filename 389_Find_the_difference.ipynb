```python
# Que- You are given two strings s and t. String t is generated by random shuffling string s and then add one more letter at a random position.
# Return the letter that was added to t.

# Approach-1(Hashmap)
from collections import defaultdict
def findthedifference(s,t):
    hashmap=defaultdict(int)
    for char1 in s:
        hashmap[char1]+=1
    for char2 in t:
        hashmap[char2]-=1
    for char3 in hashmap:
        if hashmap[char3]==-1:
            return char3                

s1="abcd" 
t1="abcde"
r1=findthedifference(s1,t1)
print(r1)


# T.C=O(n) S.C=O(n)
```

```python
# Que- You are given two strings s and t. String t is generated by random shuffling string s and then add one more letter at a random position.
# Return the letter that was added to t.

# Approach-2(ASCII conversion)

def findthedifference(s,t):
    summ1=0
    for char in s:
        summ1+=ord(char)
    summ2=0   
    for char in t:
        summ2+=ord(char)
    return chr(summ2-summ1)    

s1="abcd" 
t1="abcde"
r2=findthedifference(s1,t1)
print(r2)

# T.C=O(n) S.C=O(1)
```

```python
# Que- You are given two strings s and t. String t is generated by random shuffling string s and then add one more 
# letter at a random position.Return the letter that was added to t.

# Approach-3(Using XOR)

def findthedifference(s,t):
    XOR=0
    for char in s:
        inttt=ord(char)
        XOR=XOR^inttt
    for char in t:
        inttt2=ord(char)
        XOR=XOR^inttt2
    return chr(XOR)        
    

s1="abcd" 
t1="abcde"
r3=findthedifference(s1,t1)
print(r3)

# T.C=O(n) and S.C=O(1)
```
